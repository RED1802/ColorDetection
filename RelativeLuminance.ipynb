```python
import matplotlib.pyplot as plt
import numpy as np

# Define the relative luminance formula
def relative_luminance(r, g, b):
    return 0.2126 * r + 0.7152 * g + 0.0722 * b

# Create a gradient of RGB values
gradient_size = 100
r_values = np.linspace(0, 1, gradient_size)
g_values = np.linspace(0, 1, gradient_size)

# Create a 2D grid of RGB values
rg_values = np.array(np.meshgrid(r_values, g_values)).T.reshape(-1, 2)

# Calculate relative luminance for each RGB value
y_values = [relative_luminance(r, g, 0) for r, g in rg_values]

# Plot the RGB gradient and corresponding relative luminance values
fig, ax = plt.subplots(1, 1, figsize=(10, 6), dpi=120)

# Relative Luminance
ax.plot(rg_values[:, 0], y_values, color='black', label='Relative Luminance', linewidth=2)
ax.plot(rg_values[:, 0], 0.2126 * rg_values[:, 0], '--', label='0.2126 * Red', color='red', linewidth=2)
ax.plot(rg_values[:, 0], 0.7152 * rg_values[:, 0], '--', label='0.7152 * Green', color='green', linewidth=2)
ax.plot(rg_values[:, 0], 0.0722 * rg_values[:, 0], '--', label='0.0722 * Blue', color='blue', linewidth=2)

ax.set_title('Relative Luminance and Color Channel Contributions')
ax.set_xlabel('RGB Value')
ax.set_ylabel('Y (Relative Luminance)')
ax.legend()
ax.grid(True)

plt.tight_layout()
plt.show()
```
